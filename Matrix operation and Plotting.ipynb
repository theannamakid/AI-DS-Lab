```python
#Matrix Operations
import numpy as np
mat1=np.arange(1,10).reshape(3,3)
mat2=np.arange(11,20).reshape(3,3)
print(f"Matrix 1:\n{mat1}")
print(f"Matrix 2:\n{mat2}")
print(f"Sum of 2 matrices:\n{np.add(mat1,mat2)}")
print(f"Difference of 2 matrices:\n{np.subtract(mat2,mat1)}")
print(f"Product of 2 matrices:\n{np.dot(mat1,mat2)}")
print(f"Transpose matrix 1:\n{np.transpose(mat1)}")
```

```python
#Graph Plotting using matplotlib
import matplotlib.pyplot as plt
x=np.arange(0,10)
y=x**4
fig,ax=plt.subplots(2,2)
ax[0, 0].set_title("Normal Plot")
ax[0,0].plot(x,y)
y=[7,5,2,5,9,1,2,6,6,9]
ax[0, 1].set_title("Stem Plot")
ax[0,1].stem(x,y)
ax[1, 0].set_title("Bar Graph")
ax[1,0].bar(x,y)
ax[1, 1].set_title("Scatter Plot")
ax[1,1].scatter(x,y)
fig.tight_layout()
plt.show()
```

```python
import seaborn as sb
y=4*x**2
graph=sb.stripplot(x,y)
graph.set(xlabel='X-Axis',ylabel="Y-Axis")

plt.title("Plotting using Seaborn")
plt.show()
```
