```python
import statistics as st
import numpy as np
l=[1,7,1,3,4,3,5,5,5,6,7]
s=np.sort(l)
print(s)
print("mean = ",st.mean(l))
print("median = ",st.median(s))
print("mode = ",st.mode(l))
print("standard deviation = ",st.stdev(l))
print("percentile = ",np.percentile(l,50))
```

```python
#NORMAL DISTRIBUTION GRAPH

from scipy.stats import norm
import matplotlib.pyplot as plt
data = np.arange (1,10,0.20)
pdf=norm.pdf(data,loc=5.3,scale=1)
plt.plot(data,pdf)
plt.xlabel("Height")
plt.ylabel("Probability Density")
plt.show()
```

```python
#CDF

x=np.linspace(-4,4,1000)
y=norm.cdf(x)
plt.plot(x,y)
```
